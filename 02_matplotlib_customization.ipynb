# Matplotlib Customization – Interview-Oriented Revision

This notebook focuses on improving the appearance and readability of charts.

Topics covered:
- Colors
- Markers
- Line styles
- Figure size
- Grid
- Legends
- Axis limits
- Saving figures

Customization helps transform basic charts into clear analytical visuals.

```python
import matplotlib.pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [20000, 25000, 18000, 30000, 28000]

profit = [5000, 7000, 4000, 9000, 8500]
```

```python
# Basic Plot (Baseline)
fig, ax = plt.subplots()
ax.plot(months, sales)
plt.show()
```

```python
# Colors and Markers
fig, ax = plt.subplots()

ax.plot(months, sales, color="blue", marker="o")

ax.set_title("Monthly Sales")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue")

plt.show()
```

```python
# Line Style and Width
fig, ax = plt.subplots()

ax.plot(months, sales, color="green", linestyle="--", linewidth=2, marker="o")

ax.set_title("Monthly Sales Trend")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")

plt.show()
```

```python
# Grid and Figure Size
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(months, sales, marker="o")

ax.set_title("Sales Trend")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")

ax.grid(True)

plt.show()
```

```python
# Multiple Lines with Legend
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(months, sales, marker="o", label="Sales")
ax.plot(months, profit, marker="s", label="Profit")

ax.set_title("Sales vs Profit")
ax.set_xlabel("Month")
ax.set_ylabel("Amount")

ax.legend()

plt.show()
```

```python
# Axis Limits
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(months, sales, marker="o")

ax.set_ylim(0, 35000)

ax.set_title("Sales with Controlled Axis")

plt.show()
```

```python
# Saving Figures
fig, ax = plt.subplots()

ax.plot(months, sales, marker="o")

ax.set_title("02.Sales Chart")

plt.savefig(r"C:\Users\tanma\Desktop\DA Projects\py libs\matplotlib-seaborn\images\02.sales_chart.png")

plt.show()
```
