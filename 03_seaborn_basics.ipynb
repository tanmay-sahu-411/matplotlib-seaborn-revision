# Seaborn Basics – Interview-Oriented Revision

Seaborn is a statistical data visualization library built on top of Matplotlib.

It provides:
- Better default styles
- Simpler syntax
- Built-in statistical plots

Common uses:
- Distribution analysis
- Category comparison
- Relationship analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Why Use Seaborn?

Compared to Matplotlib, Seaborn provides:

- Better default visual styling
- Built-in support for statistical plots
- Integration with pandas DataFrames
- Simpler syntax for common chart types

Seaborn is often used during exploratory data analysis (EDA).

```python
import pandas as pd

data = {
    "product": ["Laptop", "Phone", "Tablet", "Monitor", "Laptop", "Phone", "Tablet"],
    "sales": [200, 300, 150, 180, 220, 320, 170],
    "price": [800, 500, 300, 250, 820, 520, 310]
}

df = pd.DataFrame(data)

df
```

```python
# Bar Plot (Category Comparison)
sns.barplot(x="product", y="sales", data=df)
plt.title("Average Sales by Product")
plt.show()
```

### Interpretation

This chart compares average sales across product categories.

Bar plots are useful when:
- Comparing categories
- Showing aggregated values
- Identifying best performing products

```python
# Count Plot
sns.countplot(x="product", data=df)
plt.title("Number of Records per Product")
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(x="price", y="sales", data=df)
plt.title("Price vs Sales Relationship")
plt.show()
```

```python
# Line Plot
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [200, 250, 220, 300, 280]

sns.lineplot(x=months, y=sales)

plt.title("Monthly Sales Trend")

plt.show()
```

```python
# Style Themes
sns.set_style("darkgrid")

sns.barplot(x="product", y="sales", data=df)

plt.title("Sales by Product")

plt.show()
```
