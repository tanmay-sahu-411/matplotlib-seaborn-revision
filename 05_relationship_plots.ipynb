# Relationship Plots – Interview-Oriented Revision

Relationship plots help analyze how two or more variables are related.

They are used to:
- Identify correlation
- Detect patterns
- Spot outliers

Common plots:
- Scatter Plot
- Regression Plot
- Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    "ad_spend": [10, 20, 30, 40, 50],
    "revenue": [100, 180, 260, 310, 400],
    "customers": [50, 80, 120, 150, 200]
}
df = pd.DataFrame(data)
df
```

### Scatter Plot

Used to analyze relationship between two numerical variables.

Insights:
- Positive relationship → both increase together
- Negative relationship → one increases, other decreases
- No pattern → weak/no relationship

### Interview Question
How do you identify correlation using a scatter plot?

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x="ad_spend", y="revenue", data=df)
plt.title("Ad Spend vs Revenue")

plt.show()
```

### Regression Plot

Adds a best-fit line to the scatter plot.

Helps:
- Understand trend direction
- Estimate relationship strength

### Analyst Insight
If revenue increases with ad spend → marketing is effective.

```python
plt.figure(figsize=(8,4))
sns.regplot(x="ad_spend", y="revenue", data=df)
plt.title("Ad Spend vs Revenue (Regression Line)")

plt.show()
```

### Pair Plot

Shows relationships between all numerical variables.

Useful for:
- Quick EDA
- Finding correlations
- Detecting patterns across multiple features

### Interview Question
When would you use a pair plot instead of multiple scatter plots?

```python
sns.pairplot(df)

plt.show()
```

## Summary

In this notebook we covered:

These plots are essential for understanding relationships in data during EDA.
