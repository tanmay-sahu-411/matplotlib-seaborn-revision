# Correlation Heatmaps – Interview-Oriented Revision

Correlation heatmaps help visualize the strength and direction of relationships between numerical variables.

They are useful for:
- Finding strongly related features
- Detecting multicollinearity
- Identifying patterns quickly in EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    "ad_spend": [10, 20, 30, 40, 50],
    "revenue": [100, 180, 260, 310, 400],
    "customers": [50, 80, 120, 150, 200],
    "profit": [20, 40, 70, 90, 120]
}

df = pd.DataFrame(data)

df
```

## Correlation Matrix

A correlation matrix shows the relationship between all numerical columns in a dataset.

Correlation values range from:
- **+1** → strong positive relationship
- **0** → no relationship
- **-1** → strong negative relationship

### Insights
- High positive correlation means both variables increase together
- High negative correlation means one increases while the other decreases
- Values near zero suggest weak or no linear relationship

### Interview Question
Why is correlation useful before building a model or doing analysis?

```python
df.corr()
```

## Heatmap

A heatmap is a visual representation of the correlation matrix.

It helps analysts:
- Spot strong relationships quickly
- Compare multiple variables at once
- Identify possible redundancy between features

### Analyst Insight
If two variables are highly correlated, they may be carrying similar information.

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr())
plt.title("Correlation Heatmap")

plt.show()
```

## Annotated Heatmap

An annotated heatmap displays the actual correlation values inside each cell.

This improves readability and makes interpretation easier during EDA or presentations.

### Interview Question
Why is an annotated heatmap more useful than a plain heatmap?

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Annotated Correlation Heatmap")

plt.show()
```

## Interpretation

From the heatmap, we can observe which variables move together.

For example:
- If ad spend and revenue have a high positive correlation, it suggests that higher ad spend is associated with higher revenue.
- If customers and profit are also strongly correlated, customer growth may be contributing to profitability.

### Important Note
Correlation does not mean causation.
A strong relationship does not automatically prove that one variable causes the other.
