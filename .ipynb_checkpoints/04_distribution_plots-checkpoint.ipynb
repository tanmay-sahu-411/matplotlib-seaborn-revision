# Distribution Plots – Interview-Oriented Revision

Distribution plots help analysts understand how numerical data is spread.

They help identify:
- Data concentration
- Skewness
- Outliers
- Spread of values

Common plots used:
- Histogram
- KDE Plot
- Box Plot
- Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    "marks": [45, 56, 67, 45, 78, 89, 90, 56, 67, 72, 80, 60, 49, 52]
}

df = pd.DataFrame(data)
df
```

### Histogram

A histogram shows how values are distributed across ranges.

Key insights:
- Where most values are concentrated
- Whether distribution is symmetric or skewed
- Data spread

```python
sns.histplot(df["marks"], bins=5)
plt.title("Marks Distribution")
plt.show()
```

### KDE Plot

KDE (Kernel Density Estimate) shows a smooth probability density curve.

It helps visualize:
- Data distribution shape
- Peaks in the data

```python
sns.kdeplot(df["marks"], fill=True)
plt.title("Marks Density Curve")
plt.show()
```

### Box Plot

A box plot summarizes distribution using:

- Minimum
- Q1 (25th percentile)
- Median
- Q3 (75th percentile)
- Maximum

Points outside the whiskers indicate potential outliers.

```python
sns.boxplot(x=df["marks"])
plt.title("Marks Box Plot")
plt.show()
```

### Violin Plot

A violin plot combines features of:
- Box plot
- KDE plot

It shows both:
- Distribution shape
- Summary statistics

```python
sns.violinplot(x=df["marks"])
plt.title("Marks Distribution (Violin Plot)")
plt.show()
```

## Summary

Distribution plots help analysts understand the spread and shape of numerical data.

We revised:
- Histogram for frequency distribution
- KDE for density estimation
- Box plot for statistical summary
- Violin plot for combined distribution view

These plots are widely used during exploratory data analysis (EDA).
