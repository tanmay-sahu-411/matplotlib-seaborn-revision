```python
import matplotlib.pyplot as plt
```

## What is Matplotlib?

Matplotlib is a Python data visualization library used to create static, animated, and interactive charts.

It is commonly used for:
- Trend analysis
- Category comparison
- Distribution analysis
- Relationship analysis

In interviews, Matplotlib is often treated as the foundational plotting library, while Seaborn is considered a higher-level interface built on top of it.

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [20000, 25000, 18000, 30000, 28000]

categories = ["Laptop", "Phone", "Tablet", "Monitor"]
orders = [120, 200, 90, 140]

marks = [45, 56, 67, 45, 78, 89, 90, 56, 67, 72, 80, 60, 49, 52]

ad_spend = [10, 20, 30, 40, 50]
revenue = [100, 180, 260, 310, 400]
```

## 1. Line Plot

A line plot is best used to show trends over time or ordered sequence data.

Common business use cases:
- Monthly sales trend
- Website traffic trend
- Revenue trend over quarters

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(months, sales, marker="o")
ax.set_title("Monthly Sales Trend")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
ax.grid(True)
plt.show()
```

### Interpretation
This chart shows how sales changed from January to May.

A line plot helps identify:
- Upward or downward trend
- Sudden drop or spike
- Overall movement over time

## 2. Bar Plot

A bar plot is used to compare values across categories.

Common business use cases:
- Sales by product
- Orders by region
- Revenue by store

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(categories, orders)
ax.set_title("Orders by Product Category")
ax.set_xlabel("Category")
ax.set_ylabel("Number of Orders")
plt.show()
```

### Interpretation
This chart compares order volume across product categories.

A bar chart makes it easy to compare category-level performance.

### Interview Question
Why is a bar chart better than a line plot for category comparison?

## 3. Histogram

A histogram shows the distribution of numerical data by grouping values into ranges (bins).

Common business use cases:
- Customer age distribution
- Salary distribution
- Exam score distribution

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(marks, bins=5)
ax.set_title("Distribution of Marks")
ax.set_xlabel("Marks Range")
ax.set_ylabel("Frequency")
plt.show()
```

### Interpretation
This chart shows how marks are distributed across value ranges.

A histogram helps identify:
- Concentration of values
- Spread of the data
- Possible skewness

### Interview Question
What is the role of bins in a histogram?

## 4. Scatter Plot

A scatter plot is used to study the relationship between two numerical variables.

Common business use cases:
- Ad spend vs revenue
- Price vs demand
- Experience vs salary

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(ad_spend, revenue)
ax.set_title("Ad Spend vs Revenue")
ax.set_xlabel("Ad Spend")
ax.set_ylabel("Revenue")
plt.show()
```

### Interpretation
This chart helps check whether higher ad spend is associated with higher revenue.

A scatter plot is useful for spotting:
- Positive or negative relationship
- Clusters
- Outliers

### Interview Question
What kind of patterns can you detect using a scatter plot?

## 5. Subplots

Subplots allow multiple charts to be displayed in the same figure.

This is useful when:
- Comparing multiple views together
- Creating dashboard-like layouts
- Saving space in reports

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(months, sales, marker="o")
axes[0].set_title("Sales Trend")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Sales")
axes[0].grid(True)

axes[1].bar(categories, orders)
axes[1].set_title("Orders by Category")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Orders")

plt.tight_layout()
plt.show()
```

### Interpretation
Subplots help place multiple analyses side by side.

In business reporting, this is useful when we want to show:
- Trend + category comparison
- Distribution + outlier view
- Revenue + profit view
